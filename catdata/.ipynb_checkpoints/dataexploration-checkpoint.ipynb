```python
m233p = pd.read_csv('m233p.ascd', header=0, delim_whitespace=True)
```

```python
#fig,ax = plt.subplots()



plt.figure(dpi=150)
plt.scatter(m233p[m233p['ig']==-9]['g'],m233p[m233p['ig']==-9]['i'], s=0.2)
plt.scatter(m233p[m233p['ig']==-2]['g'],m233p[m233p['ig']==-2]['i'], s=2)
plt.scatter(m233p[m233p['ig']==-1]['g'],m233p[m233p['ig']==-1]['i'], s=0.2)
plt.scatter(m233p[m233p['ig']==0]['g'],m233p[m233p['ig']==0]['i'], s=0.5)
plt.scatter(m233p[m233p['ig']==1]['g'],m233p[m233p['ig']==1]['i'], s=0.5)
plt.grid()
plt.title("g - i (coloured by ig)")
plt.ylabel('i')
plt.xlabel('g')
plt.legend(['-9','-2','-1','0','1'],loc='upper right')
```
