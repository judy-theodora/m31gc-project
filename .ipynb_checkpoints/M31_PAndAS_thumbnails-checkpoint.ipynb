```python
# to get  data from a FITS table
from astropy.io import fits
```

```python
# from https://gist.github.com/vterron/24d904f4711006b07997

import argparse
import astropy.io.fits
import os


# function to take a multi-extension and returnd the 'n th' extension.
def extract_extension(input_path, output_path, n):
    with astropy.io.fits.open(input_path, mode='readonly') as hdu:
        primary_header   = hdu[0].header
        extension_data   = hdu[n].data
        extension_header = hdu[n].header
        extension_header += primary_header
    astropy.io.fits.writeto(output_path, extension_data, extension_header,
                            output_verify='fix',overwrite=True)
```

```python
# extract a all CCD images from one MEF image 
pointing = 'm249_g'
inpath = '/Users/ahuxor/Astro_projects/GC_detection_in_images/PandAS_data/'+pointing+'.fit'
for n in range(1,36):
    outpath = '/Users/ahuxor/Astro_projects/GC_detection_in_images/PandAS_data/'+pointing+'_ccd'+str(n)+'.fit'
    extract_extension(inpath,outpath,n)
    
# and now convert to NBconvert as once images created no need to run again.
```

```python
from astropy.coordinates import SkyCoord
from astropy.wcs import WCS

fits_image_filename = '/Users/ahuxor/Astro_projects/GC_detection_in_images/PandAS_data/m249_g_ccd17.fit'
hdulist = fits.open(fits_image_filename)
print(hdulist.info())
# see https://notebook.community/darioflute/CS4A/Lecture-astronomy

header = hdulist['PRIMARY'].header
data = hdulist['PRIMARY'].data
hdulist.close()

wcs = WCS(header)

from astropy.nddata import Cutout2D

from astropy.coordinates import SkyCoord
# use coordinate of the centre of one of the Wang clusters
c0 = SkyCoord('00h37m55.96s','+41d18m35.18s ',frame='icrs')
print(c0)

center = wcs.all_world2pix(c0.ra,c0.dec,0)

print (center)

size=400
cutout = Cutout2D(data, center, size, wcs=wcs)

import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], 
                  projection=cutout.wcs)
#ax = plt.subplot(projection=wcs)
ax.set_xlabel('RA')
ax.set_ylabel('Dec')
ax.imshow(cutout.data, cmap='gray',origin='lower',vmin=700, vmax=1000)
ra = ax.coords[0]
ra.set_major_formatter('hh:mm:ss')
dec = ax.coords[1]
dec.set_major_formatter('dd:mm:ss');
# see https://openastronomy.org/rcsc18/chapters/12-images-and-visualisation/01-images
```
